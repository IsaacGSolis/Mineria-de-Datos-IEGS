Parte 3: Filtrado y Selección de Datos
1. Filtrar con múltiples condiciones: Filtra las filas que cumplan con múltiples condiciones.

```python
df_filtrado=df[df["edad"]>=18]
df_filtrado
```

```python
df_filtrado=df[(df["edad"]>=18)&(df["edad"]<40)]
df_filtrado
```

2. Seleccionar columnas: Selecciona solo ciertas columnas para un nuevo DataFrame.

```python
df=df[["edad","sexo"]]
df
```

Parte 4: Operaciones y Agrupación
1. Calcula la suma, media de una columna.

```python
df["edad"].sum()
```

```python
df["edad"].mean()
```

```python
df["edad"].std()
```

2. Agrupa los datos por una columna y calcula la media de otra columna para cada grupo.

```python
print(df.groupby("sexo")["edad"].mean())
```

3. Crea una tabla pivote para resumir los datos.

```python
tabla_piv=pd.pivot_table(df,values="edad",index="sexo",aggfunc="mean")
tabla_piv
```

Ejercicio 5: Manipulación de Datos
1. Ordenar datos: Ordena el DataFrame por una columna específica.

```python
df_ordenado=df.sort_values(by="edad",ascending=True)
df_ordenado
```

Ejercicio 6: Guardar Resultados
