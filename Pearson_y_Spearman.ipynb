```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("datos_estudiantes_decision.csv")
df.info()
```

```python
df.head()
```

```python
#Pearson
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="horas_estudio",y="calificacion")
plt.title("Horas de Estudio vs Calificacion")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificacion")
plt.show()
```

```python
#Spearman
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="horas_sueno",y="nivel_estres")
plt.title("Horas de sueño vs Nivel de estres")
plt.xlabel("Horas de sueño")
plt.ylabel("Nivel de estres")
plt.show()
```

```python
#Spearman
plt.figure(figsize=(6,4))
plt.scatter(data=df, x="uso_redes",y="calificacion")
plt.title("Uso de redes vs Calificaciones")
plt.xlabel("Uso de redes")
plt.ylabel("Calificacion")
plt.show()
```

```python
#Pearson para las 3
pearson_estudio_calf=df['horas_estudio'].corr(df['calificacion'],method='pearson')
pearson_sueno_estres=df['horas_sueno'].corr(df['nivel_estres'],method='pearson')
pearson_redes_calf=df['uso_redes'].corr(df['calificacion'],method='pearson')
print("Pearson:\n")
print(f"Horas de estudio vs calificacion: ",pearson_estudio_calf)
print(f"Horas de sueño vs nivel de estres: ",pearson_sueno_estres)
print(f"Uso de redes vs calificacion: ",pearson_redes_calf)
```

```python
#Spearman para las 3
pearson_estudio_calf=df['horas_estudio'].corr(df['calificacion'],method='spearman')
pearson_sueno_estres=df['horas_sueno'].corr(df['nivel_estres'],method='spearman')
pearson_redes_calf=df['uso_redes'].corr(df['calificacion'],method='spearman')
print("Spearman:\n")
print(f"Horas de estudio vs calificacion: ",pearson_estudio_calf)
print(f"Horas de sueño vs nivel de estres: ",pearson_sueno_estres)
print(f"Uso de redes vs calificacion: ",pearson_redes_calf)
```

**¿Coinciden Pearson y Spearman?**

No coinciden exactamente, ya que pearson mide relacion lineal y spearman mide relacion monotona, pero los resultados indican que coinciden en la direccion (positiva o negativa) que representa la grafica.

**¿Cambiarías tu decisión inicial?**
