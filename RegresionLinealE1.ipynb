```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split #Para dividir los datos en train/test
from sklearn.metrics import mean_squared_error, r2_score #Metricas de evaluación
```

```python
datos=pd.read_csv("PublicidadVentas.csv")
datos.head()
```

```python
datos.info()
```

```python
datos.describe()
```

```python
X = datos[['Inversion']].values
Y = datos[['Ventas']].values
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
modelo=LinearRegression()
modelo.fit(X_train,Y_train)
```

```python
y_pred=modelo.predict(X_test)
```

```python
mse=mean_squared_error(Y_test,y_pred)
r2=r2_score(Y_test,y_pred)
```

```python
print("Pendiente:",modelo.coef_[0][0])
print("Intercepto:",modelo.intercept_[0])
print("MSE:",mse)
print("R2:",r2)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, Y, color="blue", label="Datos reales", alpha=0.6)
plt.plot(X, modelo.predict(X), color="red", label="Regresión lineal")
plt.title("Inversión en publicidad vs Ventas")
plt.xlabel("Inversión en publicidad (miles de dólares)")
plt.ylabel("Unidades vendidas")
plt.grid(True)
plt.legend()
plt.show()
```

```python
inv_ejemplo=np.array([[1],[3],[5]])

for inversion in inv_ejemplo:
  prediccion = modelo.predict(inversion.reshape(-1,1))
  print(f"Prediccion para {inversion[0]} invertido: {prediccion[0][0]:.2f} ")
```
